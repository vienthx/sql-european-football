## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
#write your query
#Thông tin về league lấy từ bảng LEAGUE
#Thông tin về country lấy từ bảng COUNTRY
match1 = pd.read_sql("""SELECT M.MATCH_API_ID, M.HOME_TEAM_API_ID, M.AWAY_TEAM_API_ID, L.NAME, C.NAME
                        FROM MATCH M
                        LEFT JOIN COUNTRY C ON M.COUNTRY_ID = C.ID
                        LEFT JOIN LEAGUE L ON L.ID = M.LEAGUE_ID;""",conn)
match1.tail()
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
#write your query
#Số lượng trận đấu dùng COUNT, GROUP BY theo giải đấu
count_match = pd.read_sql("""SELECT L.NAME AS LEAGUE, COUNT(M.MATCH_API_ID) as COUNT FROM MATCH M, LEAGUE L
                            WHERE M.LEAGUE_ID = L.ID
                            GROUP BY M.LEAGUE_ID
                            ORDER BY COUNT DESC;""",conn)
count_match
```

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất

```python
#write your query
#SUM số bàn thắng sân nhà, chung khóa với bảng team tại team_api_id
top_20_goal_hteam = pd.read_sql("""SELECT T.TEAM_LONG_NAME AS TEAM, SUM(M.HOME_TEAM_GOAL) AS GOAL
                                FROM TEAM T, MATCH M
                                WHERE T.TEAM_API_ID = M.HOME_TEAM_API_ID
                                GROUP BY M.HOME_TEAM_API_ID
                                ORDER BY GOAL DESC
                                LIMIT 20;""", conn)
top_20_goal_hteam
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
#write your query
#Tổng số bàn thắng sân nhà theo đội bóng
#Tổng số bàn thằng sân khách theo đội bóng
#Tống số tất cả bàn thắng
team_goal = pd.read_sql(""" SELECT NAME, SUM(GOAL) FROM
                            (SELECT T.TEAM_LONG_NAME AS NAME, SUM(M.HOME_TEAM_GOAL) AS GOAL
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.HOME_TEAM_API_ID
                            GROUP BY NAME
                            UNION
                            SELECT T.TEAM_LONG_NAME AS NAME, SUM(M.AWAY_TEAM_GOAL) AS GOAL
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.AWAY_TEAM_API_ID
                            GROUP BY NAME)
                            GROUP BY NAME
                            ORDER BY GOAL DESC""",conn)

team_goal
# SUM(M.HOME_TEAM_GOAL) AS AW_GOAL
#SUM(M.AWAY_TEAM_GOAL) AS HT_GOAL,
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
#write your query
#Tổng số trận đấu tham gia trên cả sân nhà và sân khách
#Mỗi trận đấu sẽ tương ứng với một dòng, thay vì tính tổng goal như câu 12 thì đếm số lần xuất hiện
count_match = pd.read_sql("""SELECT NAME, SUM(COUNT_MATCH) AS COUNT_MATCH FROM
                            (SELECT T.TEAM_LONG_NAME AS NAME, COUNT(M.HOME_TEAM_GOAL) AS COUNT_MATCH
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.HOME_TEAM_API_ID
                            GROUP BY NAME
                            UNION
                            SELECT T.TEAM_LONG_NAME AS NAME, COUNT(M.AWAY_TEAM_GOAL) AS COUNT_MATCH
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.AWAY_TEAM_API_ID
                            GROUP BY NAME)
                            GROUP BY NAME
                            ORDER BY COUNT_MATCH DESC
                            LIMIT 20""",conn)
count_match
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu đội nhà có bàn thắng > bàn thắng của đội khách, đội nhà là "giành chiến thắng" trong trận đấu
- Nếu đội nhà có số bàn thắng < bàn thắng của đội khách, đội nhà bị "thua" trong trận đấu này
- Nếu độ nhà có bàn thắng = bàn thắng của đội khách, đội là "hòa" trong trận đấu này
- Sử dụng câu lệnh "case when"


```python
#write your query
count_WLD = pd.read_sql("""SELECT T.TEAM_LONG_NAME AS NAME,
                            COUNT(CASE WHEN M.HOME_TEAM_GOAL > M.AWAY_TEAM_GOAL THEN 1 END) AS WIN,
                            COUNT(CASE WHEN M.HOME_TEAM_GOAL < M.AWAY_TEAM_GOAL THEN 1 END) AS LOSE,
                            COUNT(CASE WHEN M.HOME_TEAM_GOAL = M.AWAY_TEAM_GOAL THEN 1 END) AS DRAW
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.HOME_TEAM_API_ID
                            GROUP BY NAME
                            ORDER BY WIN DESC""",conn)
count_WLD
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
#write your query
#LIMIT 10
#ORDER BY SỐ TRẬN THẮNG
top_10_win = pd.read_sql("""SELECT T.TEAM_LONG_NAME AS NAME,
                            COUNT(CASE WHEN M.HOME_TEAM_GOAL > M.AWAY_TEAM_GOAL THEN 1 END) AS WIN
                            FROM TEAM T, MATCH M
                            WHERE T.TEAM_API_ID = M.HOME_TEAM_API_ID
                            GROUP BY NAME
                            ORDER BY WIN DESC
                            LIMIT 10""", conn)
top_10_win
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
#write your query
#Tổng số trận đấu COUNT(*)
#Chung khóa league_id
ratio = pd.read_sql("""SELECT L.NAME AS NAME, COUNT(M.MATCH_API_ID)*100/(SELECT COUNT(M.MATCH_API_ID) FROM MATCH M) AS RATIO
                        FROM LEAGUE L, MATCH M
                        WHERE L.ID = M.LEAGUE_ID
                        GROUP BY L.NAME""", conn)
ratio
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
#write your query
#lIỆT KÊ SỐ BÀN CỦA TỪNG GIẢI ĐẤU, CẢ BÀN THẮNG SÂN KHÁCH VÀ SÂN NHÀ
#tính tổng số bàn thắng trên sân khách, sân nhà trên tất cả giải đấu
#TÍNH TỶ LỆ %
ratio_goal_league = pd.read_sql("""SELECT LEAGUE_NAME,
                                    SUM(GOAL)*100/(SELECT SUM(M.HOME_TEAM_GOAL) + SUM(M.AWAY_TEAM_GOAL) FROM MATCH M)
                                    AS RAITO_GOAL FROM
                                    (SELECT L.NAME AS LEAGUE_NAME, SUM(M.HOME_TEAM_GOAL) AS GOAL
                                    FROM LEAGUE L, MATCH M
                                    WHERE L.ID = M.LEAGUE_ID
                                    GROUP BY LEAGUE_NAME
                                    UNION
                                    SELECT L.NAME AS LEAGUE_NAME, SUM(M.AWAY_TEAM_GOAL) AS GOAL
                                    FROM LEAGUE L, MATCH M
                                    WHERE L.ID = M.LEAGUE_ID
                                    GROUP BY LEAGUE_NAME)
                                    GROUP BY LEAGUE_NAME""",conn)
ratio_goal_league
#goal = pd.read_sql("""SELECT SUM(M.HOME_TEAM_GOAL) + SUM(M.AWAY_TEAM_GOAL) FROM MATCH M""",conn)
#goal
```

## Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải.

```python
#write your query
goal_in_league = pd.read_sql("""SELECT NAME, SEASON, SUM(GOAL) AS TOTAL_GOAL FROM
                                (SELECT L.NAME AS NAME, M.SEASON AS SEASON, SUM(M.HOME_TEAM_GOAL) AS GOAL
                                FROM LEAGUE L, MATCH M
                                WHERE L.ID = M.LEAGUE_ID
                                GROUP BY L.NAME, M.SEASON
                                UNION all
                                SELECT L.NAME AS NAME, M.SEASON AS SEASON, SUM(M.AWAY_TEAM_GOAL) AS GOAL
                                FROM LEAGUE L, MATCH M
                                WHERE L.ID = M.LEAGUE_ID
                                GROUP BY L.NAME, M.SEASON)
                                GROUP BY NAME, SEASON""",conn)
goal_in_league
```

## Câu hỏi 19: Liệt kê thông tin cầu thủ:

- Quy đổi cân nặng sang kilogram
- Quy đổi chiều cao sang mét
- Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)
- Tính tuổi của cầu thủ

```python
#write your query
player = pd.read_sql("""SELECT ID, PLAYER_API_ID, PLAYER_NAME, PLAYER_FIFA_API_ID,
                        (STRFTIME('%Y','now') - STRFTIME('%Y',BIRTHDAY)) AS AGE,
                        ROUND(WEIGHT*0.453592/((HEIGHT/100)*(HEIGHT/100)),1) AS BMI,
                        HEIGHT/100 AS HEIGHT,
                        WEIGHT/2.205 AS WEIGHT
                        FROM PLAYER""",conn)
player
```

## Câu hỏi 21: Liệt kê  cầu thủ có số trận đấu nhiều nhất

```python
#write your query
#Mỗi dòng trong player_attributes thể hiện một trận đấu, chung khóa với player_fifa_api_id trong player
top = pd.read_sql("""SELECT P.PLAYER_NAME AS NAME, COUNT(PA.PLAYER_FIFA_API_ID) AS TOTAL_MATCH
                    FROM PLAYER_ATTRIBUTES PA, PLAYER P
                    WHERE P.PLAYER_FIFA_API_ID = PA.PLAYER_FIFA_API_ID
                    GROUP BY NAME
                    ORDER BY TOTAL_MATCH DESC
                    LIMIT 10""",conn)
top
```

## Câu hỏi 22: Liệt kê các cầu thủ có tổng số điểm đánh giá lớn hơn 80

```python
#write your query
#Sau khi liệt kê, tính số lần cầu thủ được đánh giá trên 80
top_player_80 = pd.read_sql("""SELECT DISTINCT(P.PLAYER_NAME) AS NAME, COUNT(PA.OVERALL_RATING) AS TOTAL FROM PLAYER P
                                LEFT JOIN PLAYER_ATTRIBUTES PA ON PA.OVERALL_RATING > 80
                                WHERE P.PLAYER_FIFA_API_ID = PA.PLAYER_FIFA_API_ID
                                GROUP BY NAME
                                ORDER BY TOTAL DESC""",conn)
top_player_80
```
